## Trabalho Final - Programação para Dispositivos Móveis 2024

### Funcionalidades Básicas
1. **Cadastro de Usuários e Grupos:**
   - Cada usuário pode criar um grupo (conta simples) ou até cinco grupos (conta prime).
   - Usuários podem ser convidados para grupos via email ou QR Code.

2. **Cadastro de Renda e Gastos:**
   - Cadastro da renda fixa de cada participante.
   - Registro de gastos fixos e variáveis.

3. **Visualização de Dados:**
   - Gráficos mostrando a diferença entre receita e gastos pessoais.
   - Comparação de gastos compartilhados entre os membros do grupo, filtráveis por mês.

4. **Tela de Gastos Compartilhados:**
   - Lista de gastos de cada usuário referente ao mês atual.

### Tecnologias Sugeridas
1. **Frontend:**
   - Utilização do framework Flutter para desenvolvimento cross-platform (iOS e Android).

2. **Backend:**
   - Uma API RESTful, possivelmente construída com Node.js, Django ou Flask.

3. **Banco de Dados:**
   - Utilização de um banco de dados relacional como PostgreSQL, MySQL ou Hasura Graphql,
   - Ou banco de dados não-relacional como Firebase todos esses citados podem gerenciar usuários, grupos, e transações.

4. **Autenticação e Segurança:**
   - Implementação de autenticação segura (JWT, OAuth).
   - Uso de HTTPS e criptografia de dados sensíveis.

### Estrutura de Dados
- **Usuário:** ID, Nome, Email, Renda Fixa, Tipo de Conta (Simples/Prime)
- **Grupo:** ID, Nome, Membros
- **Transações:** ID, Usuário, Grupo, Tipo (Fixo/Variável), Valor, Data
